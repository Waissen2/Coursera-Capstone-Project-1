```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import numpy as np
from geopy.geocoders import Nominatim 
from pandas.io.json import json_normalize  
import folium 

# CLUSTERING RESOURCE
from sklearn.cluster import KMeans

# MATPLOTLIB RESOURCES
import matplotlib.cm as cm
import matplotlib.colors as colors
from bs4 import BeautifulSoup

print('Folium installed')
print('Libraries imported.')
```

### IMPORTING DATA FROM WIKI

```python
source = requests.get('https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&oldid=1011037969').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

## Cleaning and prepairing the data for process

```python
dfs = pd.read_html(tab)
df=dfs[0]
df.head()
```

### DATA PREPARATION AND CLEANING

```python
# CHANGE ROWS Borough WHERE 'Not assigned'
df1 = df[df.Borough!='Not assigned']

# CHANGE Neighbourhoods POSTAL CODE
df2 = df1.groupby(['Postal Code','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)

# Neighbourhoods NOT ASSIGNED for Borough
df2['Neighbourhood'] = np.where(df2['Neighbourhood'] == 'Not assigned',df2['Borough'], df2['Neighbourhood'])

df2
```

### SHAPPING AS REQUESTED

```python
df2.shape
```

### RENAMING COLUMNS

```python
df2.rename(columns={'Postal Code':'Postcode'}, inplace=True)
df2.head()
```

```python
df2.rename(columns={'Postalcode':'Postcode'}, inplace=True)
df2.head()
```

### GETTING THE CSV FILE WITH LONG/LAT NEIGHBOURHOODS 

```python
lat_lon = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon.head()
```

### JOINING IN A SINGLE TABLE

```python
lat_lon.rename(columns={'Postal Code':'Postcode'},inplace=True)
df3 = pd.merge(df2,lat_lon,on='Postcode')
df3.head()
```

# NOW THE CLUSTERING PROCESS

### SEARCHING TORONTO CONTAIN IN BOROUGH

```python
df4 = df3[df3['Borough'].str.contains('Toronto',regex=False)]
df4
```

### MAPPING

```python
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(df4['Latitude'],df4['Longitude'],df4['Borough'],df4['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

### Kmeans ON ACTION

```python
k=5
toronto_clustering = df4.drop(['Postcode','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
df4.insert(0, 'Cluster Label', kmeans.labels_)
```

```python
df4
```

### GENERATING CLUSTER MAP

```python
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# COLORING
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# MARKED ON THE MAP
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(df4['Latitude'], df4['Longitude'], df4['Neighbourhood'], df4['Cluster Label']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
